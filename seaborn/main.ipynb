```python
# scatterplot ---> it is a graph which plot in points takes two major input for x and y axis
# .scatterplot function is used to create this graph and data parameter points toward the data we are using 

sns.set_style("darkgrid")
sns.set_context('notebook')
sns.scatterplot(data=charts,x="fare",y="tip",hue="tolls",palette="Set1")
sns.despine(left=True)
```

```python
charts.head()
sns.set_style('darkgrid')
sns.set_context('poster')
sns.scatterplot(data=charts,x="color",y="tolls",palette="Set1")


```
